```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_file(filename):
    df = pd.read_csv(filename)
    
    split_points = df[df['DidSplit'] == 1]
    
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['InsertionTime_us'], label='Insertion Time', zorder=1)
    
    plt.scatter(split_points.index, split_points['InsertionTime_us'], 
                color='red', s=10, label='Node Split Occurred', zorder=2)
    
    plt.xlabel('Number of Insertions')
    plt.ylabel('Time (microseconds)')
    plt.grid(True)
    plt.legend()
    
    output_filename = f"{filename.replace(' ', '_').lower()}_with_splits.png"
    plt.savefig(output_filename)
    plt.show()
    plt.close()
```

```python
plot_file('../build/in_memory_insertion_times.csv')
```

```python
plot_file('../build/on_disk_insertion_times.csv')
```
